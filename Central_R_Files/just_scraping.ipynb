```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

print("test working")

# Variable for how many seconds to wait between actions
wait_time = 3

# Set up your Chrome
option = webdriver.ChromeOptions()
option.add_argument("--incognito")        #open incognito window

# Direct to chromedriver executable file
driver = webdriver.Chrome(executable_path="/Users/matty_sam/Dropbox/Documents/CMU/Classes/Fall 2019/Applied ML/Project/Scraping/chromedriver", chrome_options = option)
driver.set_page_load_timeout(50)         #seconds (stop if loading time > 500 seconds)

###### Web navigation part
# Navigate to Reference USA through Carnegie Library
driver.get("http://clp-ezp.carnegielibrary.org/login?url=http://www.referenceusa.com")
time.sleep(3)

# Login
lib_card_ID = "11812019909279"
PIN = "1994"

# Enter Lib card Number
lib_ID_field = driver.find_element_by_xpath('//input[contains(@name,"user")]')
lib_ID_field.send_keys(lib_card_ID)

# Enter Pin
PIN_field = driver.find_element_by_xpath('//input[contains(@name,"pass")]')
PIN_field.send_keys(PIN)

# Click login
driver.find_element_by_xpath('//input[contains(@value,"Login")]').click()
time.sleep(3)
```

#  ## I'll do all manual entry to get me to the data pages ##

### Main data page - will loop through data

```python
# Getting important variables

# Getting total number of pages
max_pages = driver.find_element_by_xpath('//span[contains(@class, "data-page-max")]').text
max_pages = ''.join(max_pages.split(','))
num_pages = int(max_pages)
print("pages: ", num_pages)

# Save current URL
main_page = driver.current_url
print("main url: ", main_page) 
```

```python
# Main loop

# Actual loop code

wait_time = 4

for i in range(975,num_pages):
    page_num = str(i)
    
    ## Select Page ##
    # Click page number slots
    driver.find_element_by_xpath('//div[contains(@class,"page click-enterkey")]').click()
    time.sleep(wait_time)

    # Enter page num
    driver.find_element_by_xpath('//div[contains(@class,"pager")]/input[contains(@class,"text")]').send_keys(page_num)
    time.sleep(0.5)
    # Press enter
    driver.find_element_by_xpath('//div[contains(@class,"pager")]/input[contains(@class,"text")]').send_keys(u'\ue007')
    time.sleep(wait_time)
    
    ## Select all items ##
    driver.find_element_by_xpath('//input[contains(@class,"action-tag-all")]').click()
    time.sleep(0.5)
    
    ## Download ##
    # Go to download page
    driver.find_element_by_xpath('//a[contains(@class,"action download action-download")]').click()
    time.sleep(wait_time)
    
    # Select detailed report
    driver.find_element_by_xpath('//input[contains(@value,"detail")]').click()
    time.sleep(0.5)
    
    # Download item
    driver.find_element_by_xpath('//div[contains(@class,"buttonWrapper")]/a/span').click()
    time.sleep(0.5)
    
    # Return to main page
    driver.get(main_page)
    time.sleep(wait_time)
    
    # Unselect all elements
    driver.find_element_by_xpath('//input[contains(@class,"action-tag-all")]').click()
    
    # Print page
    print("Page", i, "completed")
    
print("Done :)")
```
